```python
from matplotlib.pyplot import cm
import numpy as np
import osmnx as ox
import networkx as nx
import random
import pandas as pd
```

```python
# Download the road network
G = ox.graph_from_place('Bremen, Germany', network_type='drive')
```

```python
# A random Blob list was created instead of Otto's input 
randList = []
n = 50
for i in range(n):
    randList.append([random.uniform(1.5, 80.5),random.uniform(1.5, 155.5)])
ids = list(range(1,n+1))

blobList = dict(zip(ids,randList))
blobList
```

```python
# Starting and ending point of a trip
start = [8.185539352132764,38.606855578588329]
endpoints = list(blobList.values())
```

```python
# Retrieve nearest nodes
nodes = []
for i in endpoints:
    nodes.append(ox.nearest_nodes(G, start, i))
nodes[:5]
```

```python
# Compute the paths of the trip
route_list = []
for i in nodes: 
    route = nx.shortest_path(G, i[0], i[1], weight='length')
    route_list.append(route)
```

```python
# Plot the trip
rc = cm.rainbow(np.linspace(0,1,len(route_list)))
fig, ax = ox.plot_graph_routes(G,
                              route_list,route_colors=rc,edge_linewidth=1,
                              node_size=1,
                              route_linewidth=6)
```

```python
lengths = []
for i in nodes:
    length = nx.shortest_path_length(G, i[0], i[1], weight='length')
    lengths.append(length)

routes = pd.DataFrame(list(zip(blobList.keys(),list(range(1,len(lengths)+1)),lengths)),columns =['blob Id','route Id','total Distance'])
routes
```
