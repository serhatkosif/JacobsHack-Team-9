```python
from matplotlib.pyplot import cm
import numpy as np
import osmnx as ox
import networkx as nx

import random
```

```python
# Download the road network
G = ox.graph_from_place('Bremen, Germany', network_type='drive')
```

```python
ox.utils_geo.sample_points(G, 50)
```

```python
rand_list = []
n = 50
for i in range(n):
    rand_list.append([random.uniform(1.5, 55.5),random.uniform(1.5, 55.5)])
ids = list(range(1,n+1))

vegetation = dict(zip(ids,rand_list))
vegetation
```

```python
# Starting and ending point of a trip
start = [8.185539352132764,38.606855578588329]
endpoints = list(vegetation.values())
```

```python
# Retrieve nearest nodes
nodes = []
for i in endpoints:
    nodes.append(ox.nearest_nodes(G, start, i))
```

```python
# Compute the paths of the trip
route_list = []
for i in nodes: 
    route = nx.shortest_path(G, i[0], i[1], weight='length')
    route_list.append(route)
```

```python
lengths = []
for i in nodes:
    length = nx.shortest_path_length(G, i[0], i[1], weight='length')
    lengths.append(length)

print(lengths)
```
